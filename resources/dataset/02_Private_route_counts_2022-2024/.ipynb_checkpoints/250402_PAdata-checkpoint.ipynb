```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
#from keplergl import KeplerGl
```

```python
df=pd.read_csv('./data/2024.csv')
df
```

# Data Processing

```python
# 假設 df 是原始資料
# Step 1：先四欄都 round 一下（可以避免浮點誤差，如果需要的話）
df['start_lat'] = df['start_lat'].round(1)
df['start_lon'] = df['start_lon'].round(1)
df['end_lat'] = df['end_lat'].round(1)
df['end_lon'] = df['end_lon'].round(1)

# Step 2：依照四欄 groupby 並統計次數
route_counts = df.groupby(['start_lat', 'start_lon', 'end_lat', 'end_lon']).size().reset_index(name='count')

# 結果：每筆是一條獨特的路線＋它出現的次數
print(route_counts.head())
```

```python
# 建立一個方向不敏感的 key（先排序再 groupby）
df['route_key'] = df.apply(
    lambda row: tuple(sorted([
        (row['start_lat'], row['start_lon']),
        (row['end_lat'], row['end_lon'])
    ])),
    axis=1
)

route_counts = df.groupby('route_key').size().reset_index(name='count')

# 分開 key 成四欄（可選）
route_counts[['pt1_lat', 'pt1_lon']] = pd.DataFrame([x[0] for x in route_counts['route_key']], index=route_counts.index)
route_counts[['pt2_lat', 'pt2_lon']] = pd.DataFrame([x[1] for x in route_counts['route_key']], index=route_counts.index)

# 如果不需要 key 可以刪掉
route_counts = route_counts.drop(columns='route_key')
```

```python
route_counts
```

```python
route_counts.nlargest(10, 'count')
```

```python
# 取出前 N 條（次數最多的）
top_routes = route_counts.sort_values(by='count', ascending=False).head(2500)
top_routes.to_csv("top_route_counts_2024.csv", index=False)
```

# Plot

```python
# 建立地圖（用第一條作為中心）
m = folium.Map(location=[top_routes['pt1_lat'].iloc[0], top_routes['pt1_lon'].iloc[0]], zoom_start=10)

for _, row in top_routes.iterrows():
    start = (row['pt1_lat'], row['pt1_lon'])
    end = (row['pt2_lat'], row['pt2_lon'])

    folium.PolyLine(
        locations=[start, end],
        color="blue",
        weight=max(1, row['count'] / top_routes['count'].max() * 8),
        opacity=0.6
    ).add_to(m)

# 存檔
m.save("top_routes_map_2024.html")
```
